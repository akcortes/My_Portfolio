## Challenge 2

How many ways can you choose 3 people from a group of 17 for making the group projects teams?

```python
#n!/(n-r)!r!
n=[i for i in range(1,18)]
r=3
```

```python
temp=combinations(n, 3)
for i in temp:
    print(i)
```
