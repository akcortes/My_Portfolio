#### The next step is to create a function that will remove references. 

We will do this by splitting the string on the `{` character and keeping only the part before this character. Write your function below.

```python
def reference(x):
    '''
    Input: A string
    Output: The string with references removed
    
    Example:
    Input: 'the{7}'
    Output: 'the'
    '''
    
    list_2=[]
    #list3=[]
    #for y in x:
    if "{" in x:
        list_2=x.split("{",1)
        return list_2[0]
    else:
        return x
    #print(list3)
    # your code here
```

```python
def line_break(x):
    '''
    Input: A string
    Output: A list of strings split on the line break (\n) character
        
    Example:
    Input: 'the\nbeloved'
    Output: ['the', 'beloved']
    
    '''
    list_2=[]
    #list3=[]
    #for y in x:
    if "\n" in x:
        list_2=x.split("\n")
        return list_2
    else:
        return x
    # your code here
```
