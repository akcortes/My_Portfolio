```python
import json
import pandas as pd
import requests

response8 = requests.get('https://api.github.com/search/code?q=.scavengerhunt in:path+repo:ironhack-datalabs/scavenger')
data6 = pd.DataFrame(response8.json())
data6
```

```python
print(response8.text)
```

```python
def flatten(data, col_list):
    for column in col_list:
        
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
nested_columns = ['items']

flat = flatten(data6, nested_columns)
flat.head(10)

ana=flat.sort_values(by="items_name",ascending=True)



```

```python
a=ana.items_path
print(a)
print(type(a))
c=a.values.tolist()
print(c)
```
